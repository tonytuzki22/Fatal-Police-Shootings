```python
def autocorrect(user_word, valid_words, diff_function, limit):
    if user_word in valid_words:
        return user_word
    a=min(valid_words,key=lambda word: diff_function(user_word,word,limit))
    if diff_function(user_word,a,limit)>limit:
            return user_word
    else:
        return a
```
