```python
import numpy as np
import pandas as pd
```

```python
'gonna start adding other variables'
income = pd.read_csv('MedianHouseholdIncome2015.csv', encoding = "ISO-8859-1")
print(len(income))

print(income.dtypes) #shows the type of data contained in cells

print(len(income))

income['Median Income'] = income['Median Income'].str.replace(r'\(.*\)', '1')#removes anything that has a () in it, done to try to clean up data
income['Median Income'] = income['Median Income'].str.replace(r'.*\-.*', '1')
income['Median Income'] = income['Median Income'].str.replace(r'\+', '')
income['Median Income'] = income['Median Income'].str.replace(r'\,', '')
income['City'] = income['City'].str.replace(r"[\s\S]*.*(balance).*[\s\S]", '')#removes everything in a cell if it has (balance) in it,prevents double counting 
income['City'] = income['City'].str.replace(r'\(.*\)', '')#removes anything that has a parenthesis in it, done to try to clean up data
income['City'] = income['City'].str.replace(r' CDP', '')#removes anything that has a 'CDP' in it, done to try to clean up data
income['City'] = income['City'].str.replace(r' city', '')#removes anything that has a lowercase spelling of 'city' in it, done to try to clean up data
income['City'] = income['City'].str.replace(r' village', '')#removes lowercase villages
income['City'] = income['City'].str.replace(r' municipality', '')#removes municipality, mainly to deal with Anchorage
income['City'] = income['City'].str.replace(r' town', '')#removes lowercase town
income['City'] = income['City'].str.replace(r'Urban ', '')#Removes Urban , for Honolulu
income['City'] = income['City'].str.replace(r'\-.*', '')#removes everything after -, used to add more per capita counts 
income['City'] = income['City'].str.replace(r'\/.*', '')#removes everything after /, used to add more per capita counts 

type_dict = {'Geographic Area': 'str', 'City': 'str', 'Median Income': 'int64'}#done to change population values into an integer
income = income.astype(type_dict)#done to change population values into an integer

print(income.dtypes)
print(len(income))


len(income)
```
